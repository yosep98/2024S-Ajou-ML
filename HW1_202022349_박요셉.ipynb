### 문제 3-2
주어진 코드틀에 k개의 "MntToyProducts" 값을 보고 평균을 내어 결측값을 결정하는 코드를 작성하세요.
* 주의) 코드틀 이외의 부분에 코드를 작성하지 마세요!
* 주의) 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
* 주의) 코드 시작과 코드 끝 라인을 삭제하지 마세요.

```python
from tqdm.notebook import tqdm
```

```python
class Knn_imputer():
    def __init__(self, n_neighbors):
        ## 생성자로 최근접 이웃수(n_neighbors)를 입력 받습니다.
        self.k = n_neighbors

    ## knn 은 lazy learning 입니다.
    def fit(self,X):
        self.X = X

    def predict(self, X):

        for i in tqdm(range(len(X))):
            if pd.isnull(X['MntToyProducts'].iloc[i]): # Arrival Delay 열에 있는 결측값 확인

                distance = self.dist(X, X.iloc[i]) # 데이터프레임 X와 결측값이 있는 행 데이터 간의 거리 구하는 함수
                sorted_dist = sorted(distance.items(), key=lambda x: x[1]) # distance 를 sorting 하기

                ################  문제 3-2  ####################
                # k개의 "MntToyProducts" 값을 보고 평균 내어
                # 값을 결정하기

                # 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
                # 코드 시작과 코드 끝 라인을 삭제하지 마세요.
                ################ 코드 시작  #####################
                # k개의 이웃 데이터의 "MntToyProducts" 값을 모으기 위한 리스트 생성
                k_nearest_neighbors = [X.iloc[idx]['MntToyProducts'] for idx, _ in sorted_dist[:self.k]]
                # 이웃 데이터의 "MntToyProducts" 값을 모은 후, 이 값을 이용하여 평균값 계산
                mean_value = sum(k_nearest_neighbors) / len(k_nearest_neighbors)
                # 계산된 평균 값을 해당 데이터의 "MntToyProducts" 값 설정
                X['MntToyProducts'].iloc[i] = mean_value
                ################ 코드 끝  ######################
                ##############################################

        return X['MntToyProducts']

    def dist(self, X, y):
        num_train = X.shape[0]

        X = X.drop(['MntToyProducts'], axis=1)
        y = y.drop(['MntToyProducts'])
        dist_dict = {} ## {index : distance} 형태

        for i in range(num_train):

            ################  문제 3-1  ####################
            # 데이터프레임 X와 결측값이 있는 행 데이터 간의 거리 구하기
            # 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
            # 코드 시작과 코드 끝 라인을 삭제하지 마세요.
            ################ 코드 시작  #####################
            # 데이터프레임 X와 결측값이 있는 행 데이터 유클리드 거리 계산 후 딕셔너리에 저장(y=결측값 데이터행)
            dist_dict[i] = np.linalg.norm(X.iloc[i] - y)
            ################ 코드 끝  ######################
            ##############################################

        return dist_dict
```

```python
imputer = Knn_imputer(3)
imputer.fit(X)

# test set에 대해 predict를 진행합니다.
data['MntToyProducts'] = imputer.predict(X) # 오래 걸립니다
```

### 문제 5-2
**`DecisionTreeClassifier`**를 사용하여 주어진 조건에 맞게 모델을 학습하고 완성한 코드 및 출력물을 보고서에 캡처하여 업로드하세요.
* 조건 1: 최대 트리의 깊이를 `depth  = [1, 3, 5, 10, 15, 30, 20]`으로 변화에 따른 정확도의 변화를 `line plot`으로 그리고 가장 높은 성능의 `depth`를 출력하시오.
* 조건 2: train set에서 validation set을 나누세요. `train : validation = 7 : 3`

* *주의! : test set이 아닌 validation set으로 평가해야 합니다.*

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
################  문제 5-2  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################

# 결측치를 평균값으로 대체
# 결측치를 평균값으로 대체하기 위한 Imputer 객체 생성
imputer = SimpleImputer(strategy='mean')
# Train set의 결측치를 평균값으로 대체
X_train_imputed = imputer.fit_transform(X_train)
# Validation set의 결측치를 평균값으로 대체
X_val_imputed = imputer.transform(X_val)
# 최대 트리의 깊이(depth) 변화에 따른 정확도의 변화 저장할 리스트
dt_scores = []
# 최대 트리의 깊이(depth) 설정
num_dep = [1, 3, 5, 10, 15, 20, 30]
# 최대 트리의 깊이(depth)를 변화에 따른 정확도 계산
for depth in num_dep:
    # DecisionTreeClassifier 모델 생성
    tree_model = DecisionTreeClassifier(max_depth=depth, random_state=42)
    # 모델 학습
    tree_model.fit(X_train_imputed, y_train)
    # 검증 데이터에 대한 정확도 계산
    accuracy = tree_model.score(X_val_imputed, y_val)
    #정확도 담기
    dt_scores.append(accuracy)
# 가장 높은 정확도 계산
max_accuracy = max(dt_scores)
# 가장 높은 정확도를 보이는 최대 트리의 깊이(depth) 선택
best_depth = num_dep[dt_scores.index(max_accuracy)]
#정확도 출력
print("가장 높은 정확도:", max_accuracy)
# 깊이 출력
print("가장 높은 정확도를 보이는 최대 트리의 깊이:", best_depth)
################ 코드 끝  ######################
##############################################
```
