```python
# 폴더 경로 설정
source_folder = 'C:\\Users\\82108\\OneDrive\\Desktop\\2024-1\\ml\\teamproj_ml\\archive\\faces_224'  # 원본 이미지가 들어있는 폴더 경로
real_folder = 'C:\\Users\\82108\\OneDrive\\Desktop\\2024-1\\ml\\teamproj_ml\\archive\\faces_real'    
fake_folder = 'C:\\Users\\82108\\OneDrive\\Desktop\\2024-1\\ml\\teamproj_ml\\archive\\faces_fake' 

# Real 이미지 이동
for filename in real_df['videoname']:
    source_path = os.path.join(source_folder, filename)
    destination_path = os.path.join(real_folder, filename)
    if os.path.exists(source_path):  # 파일 존재 여부 확인
        shutil.move(source_path, destination_path)
    else:
        print(f"Error: {source_path} does not exist.")

# Fake 이미지 이동
for filename in fake_df['videoname']:
    source_path = os.path.join(source_folder, filename)
    destination_path = os.path.join(fake_folder, filename)
    if os.path.exists(source_path):  # 파일 존재 여부 확인
        shutil.move(source_path, destination_path)
    else:
        print(f"Error: {source_path} does not exist.")
```

```python
real_filenames = real_df['videoname'].tolist()
fake_filenames = fake_df['videoname'].tolist()

real_error_count = 0
fake_error_count = 0

missing_files_real = []
for filename in real_filenames:
    if not os.path.exists(os.path.join(real_folder, filename)):
        missing_files_real.append(filename)

missing_files_fake = []
for filename in fake_filenames:
    if not os.path.exists(os.path.join(fake_folder, filename)):
        missing_files_fake.append(filename)

if len(missing_files_real) == 0:
    print("REAL:모든 이미지 파일이 폴더에 존재합니다.")
else:
    print("REAL:다음 파일이 폴더에 존재하지 않습니다:")
    for filename in missing_files_real:
        print(filename)
        real_error_count += 1


if len(missing_files_fake) == 0:
    print("FAKE:모든 이미지 파일이 폴더에 존재합니다.")
else:
    print("FAKE:다음 파일이 폴더에 존재하지 않습니다:")
    for filename in missing_files_fake:
        print(filename)
        fake_error_count += 1
        

print(f"Real 에러 수: {real_error_count}")
print(f"Fake 에러 수: {fake_error_count}")
```
